```python
order_items_products_df.groupby("product_category_name").agg({
    "order_id": "nunique",
    "price": ["min", "max", "sum"]
})
```

#### Orders Customers - Order Payments (Merge)

```python
all_df.to_csv("main_data2.csv")
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana performa penjualan dan revenue selama beberapa bulan terakhir?

```python
monthly_orders_df = all_df.resample(rule="ME", on="order_estimated_delivery_date").agg({
    "order_id": "nunique",
    "price": "sum"
}).reset_index()
monthly_orders_df.rename(columns={
    "order_estimated_delivery_date": "order_date",
    "order_id": "order_count",
    "price": "revenue"
}, inplace=True)
monthly_orders_df.sort_values("revenue", ascending=False)
```

```python
plt.figure(figsize=(20, 10))
plt.title("Jumlah Pesanan (Bulan)")
plt.plot(
    monthly_orders_df["order_date"],
    monthly_orders_df["order_count"],
    marker="o",
    linewidth=2
)
plt.ylabel("Number of order")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.title("Jumlah Pendapatan (Bulan)")
plt.plot(
    monthly_orders_df["order_date"],
    monthly_orders_df["revenue"],
    marker="o",
    linewidth=2
)
# dapetin sumbu axis saat ini
ax = plt.gca()
# format ulang sumbu y, supaya 
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.0f}'.format(x)))

plt.show()
```

### Pertanyaan 2: Produk apa yang paling banyak dan paling sedikit terjual?

```python
byproduct_df = all_df.groupby("product_category_name").order_id.nunique()
best_byproduct_df = byproduct_df.sort_values(ascending=False).reset_index()
worst_byproduct_df = byproduct_df.sort_values(ascending=True).reset_index()
```

```python
best_byproduct_df.head()
```

```python
worst_byproduct_df.head()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 10))

colors = ["green", "lightgreen", "lightgreen", "lightgreen", "lightgreen", "lightgreen", "lightgreen", "lightgreen", "lightgreen", "lightgreen"]

sns.barplot(
    data=best_byproduct_df.head(10),
    x="order_id",
    y="product_category_name",
    hue="product_category_name",
    legend=False,
    orient="h",
    palette=colors,
    ax=ax[0]
)
ax[0].set_title("Best product", fontsize=20)
ax[0].set_xlabel(None)
ax[0].set_ylabel(None)
ax[0].tick_params(axis='x', labelsize=16)
ax[0].tick_params(axis='y', labelsize=14)

sns.barplot(
    data=worst_byproduct_df.head(10),
    x="order_id",
    y="product_category_name",
    hue="product_category_name",
    legend=False,
    orient="h",
    palette=colors,
    ax=ax[1]
)
ax[1].set_title("Worst product", fontsize=20)
ax[1].set_xlabel(None)
ax[1].set_ylabel(None)
ax[1].tick_params(axis='x', labelsize=16)
ax[1].tick_params(axis='y', labelsize=14)
ax[1].yaxis.tick_right()
ax[1].invert_xaxis()

plt.show()
```

### Pertanyaan 3: Bagaimana demografi pelanggan yang kita miliki?

```python
bycustomer_state_df = all_df.groupby("customer_state").customer_id.nunique().sort_values(ascending=False).reset_index()
bycustomer_city_df = all_df.groupby("customer_city").customer_id.nunique().sort_values(ascending=False).reset_index()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(32, 10))

sns.barplot(
    data=bycustomer_state_df.head(15),
    x="customer_state",
    y="customer_id",
    ax=ax[0]
)
ax[0].set_title("Customer group by State", fontsize=20)
ax[0].tick_params(axis="x", labelsize=16)
ax[0].tick_params(axis="y", labelsize=16)
ax[0].set_xlabel(None)
ax[0].set_ylabel(None)

sns.barplot(
    data=bycustomer_city_df.head(15),
    x="customer_city",
    y="customer_id",
    ax=ax[1]
)
ax[1].set_title("Customer group by City", fontsize=20)
ax[1].tick_params(axis="x", labelsize=16)
ax[1].tick_params(axis="y", labelsize=16)
ax[1].set_xlabel(None)
ax[1].set_ylabel(None)
ax[1].set_xticklabels(labels=bycustomer_city_df["customer_city"], rotation=45, ha='right')

plt.suptitle("CUSTOMER DEMOGRAPHIC", fontsize=20)
plt.show()
```

### Pertanyaan 4: Bagaimana kualitas rating produk berdasarkan review pelanggan?

### Pertanyaan 5, 6, 7

```python
rfm_df = all_df.groupby("customer_id", as_index=False).agg({
    "order_estimated_delivery_date": "max",
    "order_id": "nunique",
    "price": "sum"
})
rfm_df.columns = ["customer_id", "max_order_timestamp", "frequency", "monetary"]
rfm_df
```

```python
recent_date = rfm_df["max_order_timestamp"].max()
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)
```

```python
rfm_df.drop("max_order_timestamp", axis=1, inplace=True)
```

```python
rfm_df.sort_values("frequency", ascending=False).head(5)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 6))

data_frequency_df = rfm_df.sort_values("frequency", ascending=False).head(5)
sns.barplot(
    data=data_frequency_df,
    hue="customer_id",
    x="customer_id",
    legend=False,
    y="frequency",
    palette=["green", "green", "green", "green", "green"],
    ax=ax[0]
)
ax[0].set_title("by Frequency", fontsize=16)
ax[0].set_xlabel(None)
ax[0].set_ylabel(None)
ax[0].tick_params(axis="y", labelsize=15)
ax[0].set_xticks(range(len(data_frequency_df["customer_id"])))
ax[0].set_xticklabels(labels=data_frequency_df["customer_id"], rotation=30, ha='right')

data_recency_df = rfm_df.sort_values("recency", ascending=True).head(5)
sns.barplot(
    data=data_recency_df,
    hue="customer_id",
    x="customer_id",
    legend=False,
    y="recency",
    palette=["green", "green", "green", "green", "green"],
    ax=ax[1]
)
ax[1].set_title("by Recency", fontsize=16)
ax[1].set_xlabel(None)
ax[1].set_ylabel(None)
ax[1].tick_params(axis="y", labelsize=15)
ax[1].set_xticks(range(len(data_recency_df["customer_id"])))
ax[1].set_xticklabels(labels=data_recency_df["customer_id"], rotation=30, ha='right')

data_monetary_df = rfm_df.sort_values("monetary", ascending=False).head(5)
sns.barplot(
    data=data_monetary_df,
    hue="customer_id",
    x="customer_id",
    legend=False,
    y="monetary",
    palette=["green", "green", "green", "green", "green"],
    ax=ax[2]
)
ax[2].set_title("by Monetary", fontsize=16)
ax[2].set_xlabel(None)
ax[2].set_ylabel(None)
ax[2].tick_params(axis="y", labelsize=15)
ax[2].set_xticks(range(len(data_monetary_df["customer_id"])))
ax[2].set_xticklabels(labels=data_monetary_df["customer_id"], rotation=30, ha='right')

plt.show()
```
